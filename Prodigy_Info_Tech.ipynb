```python
# Importing necessary libraries:
# pandas is used to handle data:

import pandas as pd
# matplotlib library is used to creat graphs , charts :

import matplotlib.pyplot as plt
# seaborn is used for more advance graphs and charts:

import seaborn as sns
# Loadind data:

data = pd.read_csv("/content/heart.csv")
# Checking data:

# print(data.head())
# checking data's shape:

print(data.shape)
# Checking null values in data:

# print(data.isnull().sum())
# checking duplicated values in data:

print(data.duplicated().sum())
# Creating figure or chart...!

plt.figure(figsize= (6,4))    # width is 6 and height is 4 here.
# Craeting countplot for gender distribution:

sns.countplot(data = data,x = "Sex",palette = ["#FF69B4", "#87CEFA"])
# Setting title and labels for the chart:

plt.title("Sex Distribution in Heart Disease Dataset",fontsize = 14)
plt.xlabel("Sex",fontsize = 12)
plt.ylabel("Count",fontsize = 12)
# Dispalying chart.....!

plt.show()
```
